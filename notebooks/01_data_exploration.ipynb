```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
current_dir = os.getcwd()

root_dir = os.path.dirname(current_dir)
```

```python
x_train = np.load(os.path.join(root_dir, 'kmnist-train-imgs.npz'))['arr_0']
y_train = np.load(os.path.join(root_dir, 'kmnist-train-labels.npz'))['arr_0']
x_test = np.load(os.path.join(root_dir, 'kmnist-test-imgs.npz'))['arr_0']
y_test = np.load(os.path.join(root_dir, 'kmnist-test-labels.npz'))['arr_0']
```

```python
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.axis('off')
plt.show()
```

```python
print(f"Training samples: {x_train.shape[0]}")
print(f"Test samples: {x_test.shape[0]}")
print(f"Image shape: {x_train.shape[1:]}")
print(f"Number of classes: {len(np.unique(y_train))}")
```
